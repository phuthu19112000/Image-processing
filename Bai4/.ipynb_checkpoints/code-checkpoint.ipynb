1. Thêm nhieu Gaussian và muối tiêu với các thông số khác nhau vào hình ảnh bạn chọn. Đánh giá mức độ nhiễu mà bạn cho là vẫn có thể chấp nhận được để kiểm tra trực quan hình ảnh.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import skimage
```

```python
img = cv2.imread("Bac.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

def add_noise(img,mod):
    if mod == "gaussian":
        img_noise = skimage.util.random_noise(img, mode = mod, mean = 0, var = 0.01)
    elif mod == "s&p":
        img_noise = skimage.util.random_noise(img, mode = mod,amount = 0.05)
    return img_noise
```

```python
img_noise_gauss = add_noise(img,"gaussian")
plt.imshow(img_noise_gauss,cmap="gray")
plt.title("image + noise gaussian")
plt.axis("off")
```

```python
img_noise_salt = add_noise(img,"s&p")
plt.imshow(img_noise_salt,cmap="gray")
plt.title("image + noise muối tiêu")
plt.axis("off")
```

2. Áp dụng bộ lọc trung vị cho các hình ảnh bạn thu được ở trên. Thay đổi kích thước cửa sổ của bộ lọc và đánh giá mối quan hệ của nó với mức độ nhiễu.

```python
img_noise_gauss = np.array(255*img_noise_gauss, dtype = 'uint8')
img_noise_gauss = cv2.medianBlur(img_noise_gauss,3)
plt.imshow(img_noise_gauss,cmap="gray")
plt.title("Lọc trung vị cho noise gauss")
plt.axis("off")
```

```python
img_noise_salt = np.array(255*img_noise_salt, dtype = 'uint8')
img_noise_salt = cv2.medianBlur(img_noise_salt,3)
plt.imshow(img_noise_salt,cmap="gray")
plt.title("Lọc trung vị cho noise salt")
plt.axis("off")
```

3. Thực hiện thuật toán lọc Wiener, Hãy xem xét cộng nhiễu Gaussian và sau đó khôi phục lại hình ảnh ban đầu từ nhiễu.

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import color, data, restoration
from scipy.signal import convolve2d as conv2


img = cv2.imread("Bac.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
psf = np.ones((5, 5)) / 25
image = conv2(img, psf, 'same')
image += 0.3 * image.std() * np.random.standard_normal(image.shape)
deconvolved = restoration.wiener(image, psf, 1, clip=False)

plt.figure(figsize=(10,20))
plt.subplot(1,3,1)
plt.imshow(img,cmap = "gray")
plt.title("origin image")
plt.axis("off")
plt.subplot(1,3,2)
plt.imshow(image,cmap='gray')
plt.title("image+noise+blur")
plt.axis("off")
plt.subplot(1,3,3)
plt.imshow(deconvolved, cmap='gray')
plt.title("wiener")
plt.axis("off")
plt.show()
```
