Thực hiện một cách triển khai cơ bản của JPEG:

1. Chia hình ảnh thành các khối 8x8 không chồng lên nhau.
2. Tính DCT (biến đổi cosin rời rạc) của mỗi khối.
3. Lượng tử hóa từng khối. Bạn có thể làm điều này bằng cách sử dụng các bảng trong video hoặc chỉ cần chia từng hệ số cho N, làm tròn kết quả thành số nguyên gần nhất và nhân lại với N. Hãy thử các giá trị khác nhau của N.
4. Bạn cũng có thể thử bảo toàn 8 hệ số lớn nhất (trong tổng số 8x8 = 64) và chỉ cần làm tròn chúng thành số nguyên gần nhất.
5. Hình dung kết quả sau khi đảo ngược lượng tử hóa và DCT.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.colors import Normalize
import matplotlib.cm as cm
```

```python
QUANTIZATION_MATRIX = np.array([
 [16, 11, 10, 16, 24, 40, 51, 61],
 [12, 12, 14, 19, 26, 58, 60, 55],
 [14, 13, 16, 24, 40, 57, 69, 56],
 [14, 17, 22, 29, 51, 87, 80, 62],
 [18, 22, 37, 56, 68, 109, 103, 77],
 [24, 35, 55, 64, 81, 104, 113, 92],
 [49, 64, 78, 87, 103, 121, 120, 101],
 [72, 92, 95, 98, 112, 100, 103, 99]
])
```

Chiều cao và chiều rộng của các khối là B = 8. Hình ảnh được import và chuyển thành hình ảnh thang độ xám. Sau đó, hình ảnh bị cắt, sao cho chiều cao và chiều rộng của nó là bội số của kích thước khối B.:

```python
B=8 #blocksize
fn3= 'baby.png'
img1 = cv2.imread(fn3, cv2.IMREAD_GRAYSCALE)
h,w=np.array(img1.shape[:2])/B * B
print(h)
print(w)
img1=img1[:int(h),:int(w)]
```

Đối với mỗi khối trong hình ảnh, DCT được áp dụng. Hình ảnh đã biến đổi được lưu trữ trong biến Trans và nó được lưu vào tệp Transformed.jpg:

```python
blocksV=h/B
blocksH=w/B
vis0 = np.zeros((int(h),int(w)), np.float32)
Trans = np.zeros((int(h),int(w)), np.float32)
vis0[:int(h),:int(w)] = img1
for row in range(int(blocksV)):
    for col in range(int(blocksH)):
        currentblock = cv2.dct(vis0[row*B:(row+1)*B,col*B:(col+1)*B])
        Trans[row*B:(row+1)*B,col*B:(col+1)*B]=currentblock

cv2.imwrite("Transformed.jpg",Trans)
```

Để quan sát phép biến đổi DCT trông như thế nào ta chon một vùng nhỏ nhất định trên ảnh, sau đó thực hiện biến đổi DCT ta thấy phần lớn thông tin của bức ảnh được tập trung vào góc trên bên trái của matrix DCT

```python
plt.figure()
plt.subplot(1,2,1)
selectedImg=img1[row*B:(row+1)*B,col*B:(col+1)*B]
N255=Normalize(0,255)
plt.imshow(selectedImg,cmap="gray",norm=N255,interpolation='nearest')
plt.title("Image in selected Region")

plt.subplot(1,2,2)
selectedTrans=Trans[row*B:(row+1)*B,col*B:(col+1)*B]
plt.imshow(selectedTrans,cmap=cm.jet,interpolation='nearest')
plt.colorbar(shrink=0.5)
plt.title("DCT transform of selected Region")
```

Sau đó, IDCT được áp dụng để tái tạo lại hình ảnh ban đầu từ biểu diễn đã biến đổi. Hình ảnh được tái tạo được lưu trữ trong biến back0 và nó được lưu vào tệp BackTransformed.jpg:

```python
back0 = np.zeros((int(h),int(w)), np.float32)
for row in range(int(blocksV)):
    for col in range(int(blocksH)):
        currentblock = cv2.idct(Trans[row*B:(row+1)*B,col*B:(col+1)*B])
        back0[row*B:(row+1)*B,col*B:(col+1)*B]=currentblock
cv2.imwrite('BackTransformed.jpg', back0)
```

Để xác minh rằng DCT và IDCT là không mất dữ liệu, Sai số  tuyệt đối Trung bình (MAD) giữa hình ảnh gốc và hình ảnh được tái tạo sẽ được tính toán.

```python
diff=back0-img1
print(diff.max())
print(diff.min())
MAD=np.sum(np.abs(diff))/float(h*w)
print ("Mean Absolute Difference: ",MAD)
plt.figure()
plt.imshow(back0,cmap="gray")
plt.title("Backtransformed Image")
plt.axis("off")
plt.show()
```
